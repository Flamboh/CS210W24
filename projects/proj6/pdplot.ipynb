```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

df = pd.read_json("population.json")

df.head()

dot_size = [x / 1000 for x in df['pop2023']]

df.plot(x="lng", y="lat", kind="scatter", s=dot_size, color="lightgreen")
```

```python
us_states = gpd.read_file('https://raw.githubusercontent.com/PublicaMundi/MappingAPI/master/data/geojson/us-states.json')

state_name = 'Oregon'  
state_map = us_states[us_states['name'] == state_name]

fig, ax = plt.subplots(figsize=(8,6))

state_map.plot(color='lightgrey', ax=ax)
df.plot(x="lng", y="lat", kind="scatter", s=dot_size, color="green", ax=ax)

plt.show()
```
